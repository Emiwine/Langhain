### Ollama 

Ollama support embedding model , make it possible to create Rag(Retrival Augmented Generation) based application that combine text promts to existing technique

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embedding = (
    OllamaEmbeddings(model="all-minilm")  ## by default use llama2 
)
```

```python
embedding
```

```python
r1 = embedding.embed_documents(  
    [
        "Alpha is the first letter of the Greek Alphabet"
        "Beta is the second letter of the Greek Alphabet"
    ]
)
```

```python
r1[0]
```

```python
embedding.embed_query("What is the second letter of greek alphabet ")
```
